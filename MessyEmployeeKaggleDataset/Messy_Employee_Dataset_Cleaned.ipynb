```python
import pandas as pd
```

```python
# importing the csv
df = pd.read_csv(r"C:\Users\Raffy\Downloads\Compressed\Messy Employee Dataset\Messy_Employee_dataset.csv")
```

```python
df
```

```python
# Check if there is a null row in all columns
df.isna().sum()
```

```python
# fill the missing value for column Age with median
df['Age'] = df['Age'].fillna(df['Age'].median()).round().astype(int)
```

```python
# Split the Department_Region to 2 separate column name Department and Region
df[['Department', 'Region']] = df['Department_Region'].str.split('-', n=1, expand=True)
```

```python
# Standardize date column to YYYY-MM-DD format
df['Join_Date'] = pd.to_datetime(df['Join_Date'],errors = 'coerce')
```

```python
# fill the missing value for column Salary with average salary rounded by 2 decimals
df['Salary'] = df['Salary'].fillna(df['Salary'].mean()).round(2)
```

```python
# Remove the dash sign in the Phone column
df['Phone'] = df['Phone'].abs()
```

```python
# Standardize the Phone column to be 555-555-5555 format
df['Phone'] = df['Phone'].apply(lambda x: str(x))
df['Phone'] = df['Phone'].apply(lambda x: x[0:3]+'-'+x[3:6]+'-'+x[6:10])
```

```python
# Converted the Performance_Score column into numerical scale
mapping = {'Poor':1,'Average':2,'Good':3,'Excellent':4}
df['Performance_Score'] = df['Performance_Score'].map(mapping)
```

```python
# Converted the Remote_Work into a binary format
df['Remote_Work'] = df['Remote_Work'].astype(int)
```
